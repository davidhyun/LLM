# Retrieval - DocumentLoaders
Document Loaders는 다양한 형태의 문서를 RAG 전용 객체로 불러오고 이를 LangChain에 결합하기 쉬운 텍스트 형태로 반환하는 기능을 한다. 이를 통해 사용자는 txt, pdf, word, ppt, xlsx, csv 등 거의 모든 형식의 문서를 기반으로 LLM을 구동할 수 있다.

- Page_content : 문서의 내용
- Metadata : 문서의 위치, 제목, 페이지 넘버 등

## 1.URL Document Loader

```python
from langchain.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://n.news.naver.com/mnews/article/092/0002307222?sid=105")

data = loader.load() # List로 반환
print(data[0].page_content) # 모든 내용을 가져옴
```

```python
from langchain.document_loaders import UnstructuredURLLoader

urls = [
    "https://n.news.naver.com/mnews/article/092/0002307222?sid=105",
    "https://n.news.naver.com/mnews/article/052/0001944792?sid=105",
]

loader = UnstructuredURLLoader(urls=urls)
data = loader.load()
data
```

## 2.PDF Document Loader

```python
from langchain.document_loaders import PyPDFLoader

loader = PyPDFLoader('/content/dirve/MyDrive/ColabNotebooks/LLM/docs/pdf-sample.pdf')
pages = loader.load_and_split()
```

```python
print(pages[0].page_content)
print(pages[0].metadata)
```

## 3.Word Document Loader

```python
from langchain.document_loaders import Docx2txtLoader

loader = Docx2txtLoader('/content/dirve/MyDrive/ColabNotebooks/LLM/docs/word-sample.docx')
data = loader.load()
data
```

```python
print(data[0].page_content)
print(data[0].metadata)
```

## 4.CSV Document Loader


```python
from langchain.document_loaders.csv_loader import CSVLoader

loader = CSVLoader(
    file_path='/content/dirve/MyDrive/ColabNotebooks/LLM/docs/csv-sample.csv',
    csv_args={
        'delimiter': ',',
        'quotechar': '"',
        'fieldnames': ['title','company','location','link']
        }
    )

data = loader.load() # 행(row) 단위로 리스트 반환
data
```
