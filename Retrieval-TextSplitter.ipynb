## 1.CharacterTextSplitter
특정 구분자를 기준으로 텍스트를 청크로 분할하여 토큰 제한 이슈를 우회하고 Context를 학습시킬 수 있다. (max_token을 초과할 수도 있음)

```python
with open('/content/dirve/MyDrive/ColabNotebooks/LLM/docs/state_of_the_union.txt') as f:
    state_of_the_union = f.read()
state_of_the_union
```

```python
from langchain.text_splitter import CharacterTextSplitter
text_splitter = CharacterTextSplitter(
    separator = "\n\n",      # 하위 청크로 나누는 구분자
    chunk_size = 300,
    chunk_overlap = 100,     # 문맥에 대한 이해를 위해 문단 앞뒤를 얼마만큼 겹치게 할지 결정하는 파라미터
    length_function = len,   # chunk size 측정 기준
)
```

```python
texts = text_splitter.split_text(state_of_the_union)
print(texts[0])
print("-"*100)
print(texts[1])
print("-"*100)
print(texts[2]) # 세번째 chunk가 두번째 chunk의 마지막 부분을 overlap
```

```python
char_list = []
for i in range(len(texts)):
    char_list.append(len(texts[i]))
    if len(texts[i]) >= 300:
        print(len(texts[i])) # chunk size를 넘기도 함

print(char_list)
```

```python
text_splitter.create_documents([state_of_the_union]) # Document로 생성 가능
```

## 2.RecursiveCharacterTextSplitter
재귀적으로 문서를 분할한다. 문장들의 의미를 최대한 보존하는 형태로 보존할 수 있고, 다수의 chunk를 LLM에 활용함에 있어 맥락이 유지되도록하기에 용이한 방법이다.
1. **줄바꿈(`\n\n`)**을 기준으로 문서를 분할
2. chunk size가 여전히 클 경우 **문장단위(`\n`)**를 기준으로 문서를 분할
3. chunk size가 여전히 클 경우 **단어단위**로 분할


```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1000,
    chunk_overlap = 200,
    length_function = len,
)
```

```python
texts = text_splitter.create_documents([state_of_the_union])
print(texts[0].page_content)
print("-"*500)
print(texts[1].page_content)
```

```python
char_list = []
for i in range(len(texts)):
    char_list.append(len(texts[i].page_content))
    if len(texts[i].page_content) >= 1000:
        print(len(texts[i].page_content)) # chunk size를 넘기도 함

print(char_list) # 설정한 chunk_size 1000을 넘는 chunk가 없음
```

```python
from langchain.document_loaders import PyPDFLoader

loader = PyPDFLoader("/content/dirve/MyDrive/ColabNotebooks/LLM/docs/[이슈리포트 2022-2호] 혁신성장 정책금융 동향.pdf")
pages = loader.load_and_split() # Document 객체 생성
print(len(pages))
```

```python
texts = text_splitter.split_documents(pages)
print(len(texts)) # 18page로부터 31개의 chunk로 분할됨
print(texts[1].page_content)
```

## 3.EtcSplitter
- 일반적인 글로 된 문서는 모두 text splitter로 분할할 수 있으며, 대부분의 경우 커버된다.
- 그러나 코드, latex 등과 같이 컴퓨터 언어로 작성되는 문서의 경우 text splitter로 처리할 수 없으며 해당 언어를 위해 특별하게 구분하는 splitter가 필요하다.
- Python 문서를 분할하기 위해 `class`, `def`와 같이 하나의 단위로 묶이는 것을 기준으로 문서를 분할할 필요가 있다.
- 이러한 원리로 Latex, HTML, Code 등 다양한 문서도 분할할 수 있다.

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter, Language

RecursiveCharacterTextSplitter.get_separators_for_language(Language.PYTHON) # 파이썬코드 문서 분할 기준
```

```python
PYTHON_CODE = """
def hello_world():
    print("Hello, World!")

# Call the function
hello_world()
"""

python_splitter = RecursiveCharacterTextSplitter.from_language(
    language=Language.PYTHON,
    chunk_size=50,
    chunk_overlap=0,
)

python_docs = python_splitter.create_documents([PYTHON_CODE])
python_docs # 2개의 chunk로 분할
```

## 4.TokenSplitter
- 텍스트 분할의 목적은 LLM이 소화할 수 있을 정도의 텍스트만 호출하도록 만드는 것이다. 따라서 LLM이 소화할 수 있는 양으로 chunk를 제한하는 것은 LLM앱을 개발할 때 필수적인 과정이다.
- LLM은 텍스트를 받아들일 때, 정해진 토큰 이상으로 소화할 수 없게 설계되어 있다. 따라서 글을 토큰 단위로 분할한다면 최대한 많은 글을 포함하도록 chunk를 분할할 수 있다.
- 토큰이라는 것은 텍스트와 달리 Transformer에서 처리하는 방식에 따라서 그 수가 달라질 수 있다. 따라서 LLM 앱을 개발하고자 한다면 앱에 얹힐 LLM의 토큰 제한을 파악하고, 해당 LLM이 사용하는 Embedder를 기반으로 토큰 수를 계산해야 한다.
- 예를들어, OpenAI의 GPT 모델은 **tiktoken**이라는 토크나이저를 기반으로 텍스트를 토근화 한다. 따라서 tiktoken encoder를 기반으로 텍스트를 토큰화하고, 토큰 수를 기준으로 텍스트를 분할하는 것이 프로덕트 개발의 필수 요소이다.

```python
!pip install tiktoken
```

```python
import tiktoken
tokenizer = tiktoken.get_encoding("cl100k_base") # GPT모델들이 사용하는 임베딩 모델

def tiktoken_len(text):
    tokens = tokenizer.encode(text)
    return len(tokens)
```

```python
tiktoken_len(texts[1].page_content)
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=0,
    length_function = tiktoken_len # 토큰수를 기준으로 chunk 분할
)

texts = text_splitter.split_documents(pages)
print(len(texts[1].page_content)) # 글자수는 1068자
print(tiktoken_len(texts[1].page_content)) # 토큰수는 960개
```

```python
token_list = []
for i in range(len(texts)):
    token_list.append(tiktoken_len(texts[i].page_content))

print(token_list) # 1000이하로 토큰 수를 제한하였기 때문에 1000을 넘는 chunk가 없음 (토큰 제한 이슈를 피할 수 있음)
```
